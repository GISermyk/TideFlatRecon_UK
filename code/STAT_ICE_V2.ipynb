```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import glob
import re

def checlk_h5_valid(path):

    try:
        with h5py.File(path, 'r'):
            return True
    except Exception as e:
        print('bad file:', path)
        print(e)
        return False

def get_atl_group(folder_path_03, folder_path_08):

    alt03_file_list = glob.glob(os.path.join(folder_path_03, '*ATL03*.h5'))
    n = len(alt03_file_list)

    ATL_name_group = []
    for i in range(n):
        
        # find keyword (date and orbit number)
        file_path = alt03_file_list[i]
        time_match = re.search(r'ATL03_(\d{14})_', file_path)
        orbit_match = re.search(r'ATL03_\d{14}_(\d{8})_', file_path)
        Imaging_date = time_match.group(1)[:8]
        orbit_num = orbit_match.group(1)

        # search target ATL03/08 file with the same keyword (data)
        alt03_file_list_ = glob.glob(os.path.join(folder_path_03, f'*ATL03*{Imaging_date}*{orbit_num}*.h5'))
        alt08_file_list = glob.glob(os.path.join(folder_path_08, f'*ATL08*{Imaging_date}*{orbit_num}*.h5'))
        
        # Construct ATL_pair and storage in ATL_name_group( 2-D List)
        if (len(alt08_file_list) == 1) & (len(alt08_file_list) == 1):
            
            ATL_pair = [alt03_file_list_[0], alt08_file_list[0]]

            if checlk_h5_valid(alt03_file_list_[0]) & checlk_h5_valid(alt08_file_list[0]):

                ATL_name_group.append(ATL_pair)

    return ATL_name_group


def find_anomaly(indices_list):
    b = []
    if len(indices_list) <= 5:
        value = 100
    else:
        for i in range(1,len(indices_list)):
            b.append(indices_list[i] - indices_list[i-1])
            
        if max(b) > 5:
            index_max = [index for index, value in enumerate(b) if value == max(b)]
            value = indices_list[index_max[0]+1]
        else:
            value = indices_list[0]
        # value = indices_list[0]
    return value

def sea_height_cal(input, height_name, derivative_thred):
    
    if  isinstance(input, str):
        data = pd.read_csv(input)
        df = pd.DataFrame(data)
    else:
        df = input
    
    percentile, derivative, derivative_2th = [], [0], [0]
    height_TS = pd.DataFrame()
    
    if len(df) < 100:
        return 999, df, height_TS
    else:
        
        for i in range(0, 101):
            p = np.percentile(df[height_name], i)
            percentile.append(p)

        height_TS['percentile'] = percentile
        
        height_TS['percentile'][:5] = height_TS['percentile'][5]
        
        short_window, long_window = 5, 10
        height_TS['EMA5'] = height_TS['percentile'].ewm(span=short_window, adjust=False).mean()
        height_TS['EMA10'] = height_TS['percentile'].ewm(span=long_window, adjust=False).mean()
        height_TS['MACD'] = height_TS['EMA5'] - height_TS['EMA10']

        indices = [index for index, value in enumerate(height_TS['MACD']) if value > derivative_thred]
        target_index = find_anomaly(indices)
        # if indices == []:
        #     indices = [100]

        print(indices[:10], 'indices')
        print(target_index, 'target_index')
        # sea_height = percentile[target_index]
        sea_height = percentile[target_index]
        
        return  sea_height, df, height_TS 
     #, target_index

def icesat_gt_processing3_new(No ,alt03_path, alt08_path, beam, region_mask, TF_height_range, gap_distance, sea_thred, CHUNK_SIZE ,is_display = False):
    
    osgm_path = 'E:\EGM2008_Gyeonggi.tif'
    # df_sea_point_remove_list = []
    # local_strip_height_list = []
    log = []
       
    atl03_data = get_atl03_data(alt03_path, beam)
    atl08_data = read_hdf5_atl08(alt08_path, beam, False)

    if (atl03_data is None) | (atl08_data is None) :
        print(f'{alt03_path} is None')
        empty = pd.DataFrame()
        return empty, [], [], [], [], [], []
    
    add_atl08_classed_flag(alt08_path, beam, atl03_data)
    
    x_origin, y_origin, conf, lat, lon, classed_pc_flag = select_atl03_data(atl03_data, region_mask, TF_height_range)
    z = pd.DataFrame()
    z['X_Along_track'], z['height'], z['latitude'], z['longtitude'] = x_origin, y_origin, lat, lon
    z['distance'] = (z['X_Along_track'] // gap_distance) * gap_distance
    
    def filter_and_mean(group):

        M = group['height'].mean()
        S = group['height'].std()

        valid = group[(group['height'] >= M-1*S) & (group['height'] <= M+1*S)]
        return pd.Series({
            'mean_height': valid['height'].mean(),
            'mean_lon': valid['longtitude'].mean(),
            'mean_lat': valid['latitude'].mean()
        })

    re_accumulate = z.groupby('distance').apply(filter_and_mean).reset_index(drop=True)

    if osgm_path is not None and not re_accumulate.empty:
                    
        raster = rasterio.open(osgm_path)
        transformer = Transformer.from_crs(4326, raster.crs, always_xy=True)
        xs, ys = transformer.transform(re_accumulate['mean_lon'].values, re_accumulate['mean_lat'].values)
        coords = list(zip(xs, ys))
        re_accumulate['OSGM'] = [val[0] if val is not None else None for val in raster.sample(coords)]
        re_accumulate['OH'] = re_accumulate['mean_height'] - re_accumulate['OSGM']
                
        # remove land 
        mask = (re_accumulate['OH'] < 10) & (re_accumulate['OH'] > -10) 
        re_accumulate = re_accumulate[mask]
    
    empty_pf = pd.DataFrame()

    if re_accumulate.empty | len(re_accumulate) < 200:
        print('photon less than 200!')
        return empty_pf,  [], [], [], [], [], []

    #---------------------------------------------------------------------------
    # calculate global threshold and local threshold

    sea_height, df, height_TS = sea_height_cal(re_accumulate, 'OH', sea_thred)  

    if height_TS['MACD'].max() < 0.1:
        print('all track is not inundated')
        print(height_TS['MACD'].max())
        final_sea_height = 10
        ts_values, segments = [], []
    else:

        CHUNK_SIZE = CHUNK_SIZE
        MIN_LAST_CHUNK_SIZE = 200  # 定义最小的尾部块大小阈值
        total_len = len(re_accumulate)
        segments, ts_values, ranges = [], [], []

        for i in range(0, total_len, CHUNK_SIZE):

            start = i
            end = min(i + CHUNK_SIZE, total_len)
            
            # 检查下一个可能的块是否太小，如果是，则将当前块延伸到最后
            # i + CHUNK_SIZE 是下一个块的起点
            remaining_length = total_len - (i + CHUNK_SIZE) 
            
            if remaining_length > 0 and remaining_length < MIN_LAST_CHUNK_SIZE:
                end = total_len 
                ranges.append((start, end))
                break
            
            ranges.append((start, end))

        for start, end in ranges:
            chunk = re_accumulate.iloc[start:end]
            if not chunk.empty:

                sea_height_val, df_result, height_TS_sub = sea_height_cal(chunk, 'OH', sea_thred)
                
                # 将计算出的 'sea_height' 列表/数组添加到 DataFrame 中
                df_result['sea_height'] = sea_height_val

                segments.append(df_result)
                ts_values.append(height_TS_sub)
        print(50*'#')
        
        combined_df = pd.concat(segments, ignore_index=False)
        final_sea_height = combined_df['sea_height']
 
    df['Strip_No'] =  str(No) + '_' + beam
    
    #--------------------------------------------------------------------------------------------------------------  
        
    log.append('sea height list:' + str(sea_height) + ' --- ')

    # calculate Final_sea_height
        
    Final_sea_height = sea_height
    # local_strip_height_list.append(Final_sea_height)
        
    log.append('Final_sea_height :'+ str(Final_sea_height) + '\n')
            
    #---------------------------------------------------------------------------------------------------------------
    
    # Apply Final_sea_height for all beams
    # mask = df['OH'] > Final_sea_height
    df['sea_height'] = final_sea_height
    df['sea_point'] = df['OH'] <= df['sea_height']
    mask = df['OH'] > df['sea_height']

    
    print(type(mask), 'type mask', len(mask), 'mask len') 
            
    df_sea_point_remove = df[mask]
            
    if df_sea_point_remove.empty:
        print(filepath_03 + '_' + beam + ' is empty')
            
    if is_display:
            
        plt.figure(figsize=(10, 3))
        plt.plot(df['OH'], color ='gray',label = 'height')
        
        if height_TS['MACD'].max() < 0.1:
            plt.plot(df['sea_height'], color = 'green' , linestyle = '--')
        else:
            for df_result in segments:
                plt.plot(df_result['sea_height'], color = 'green' , linestyle = '--')

        plt.scatter(df.index[df['sea_point']], df['OH'][df['sea_point']], color = 'red', label = 'sea points', zorder = 0.5)
        plt.legend()
        plt.ylabel('height')
        plt.xlabel('photons')
        plt.show()
        
    if len(df) == 0:
        empty = pd.DataFrame()
        return empty, [], [], [], [], [], []
        
    return df_sea_point_remove, sea_height, height_TS, log, re_accumulate,  df, ts_values#, height_TS_11, height_TS_12
```
